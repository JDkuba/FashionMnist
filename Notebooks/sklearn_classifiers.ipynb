```
%cd '/content/drive/My Drive/Projekt/FashionMnist'

from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sn
import time

import mnist_reader
X_train, y_train = mnist_reader.load_mnist('data/fashion', kind='train')
X_test, y_test = mnist_reader.load_mnist('data/fashion', kind='t10k')
```

```
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(criterion="entropy", max_depth=50, n_estimators=100)
```

```
start = time.time()
rf_clf.fit(X_train, y_train)   # Approximately less than 2 minutes
stop = time.time()
```

```
acc = (rf_clf.predict(X_test) == y_test).sum() / len(y_test)
print(f"Accuracy: {acc}")
print(f"Training time: {stop - start}s")

plot_confusion_matrix(rf_clf, X_test, y_test, values_format="")
plt.show()
```

```
from sklearn.neural_network import MLPClassifier

mlp_clf = MLPClassifier(activation="relu", hidden_layer_sizes=[100])
```

```
start = time.time()
mlp_clf.fit(X_train, y_train) # Approximately less than 10 minutes
stop = time.time()
```

```
acc = (mlp_clf.predict(X_test) == y_test).sum() / len(y_test)
print(f"Accuracy: {acc}")
print(f"Training time: {stop - start}s")

plot_confusion_matrix(mlp_clf, X_test, y_test, values_format="")
plt.show()
```
